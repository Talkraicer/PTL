```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
data_path = "output_results/network_toy/Toy/StaticNumPass_1"
```

```python
# read data
metrics = ["passDelay","totalDelay"]
for metric in metrics:
    metric_data = os.path.join(data_path, metric+ ".pkl")
    df = pd.read_pickle(metric_data)
    demands = sorted(list(set([col[0] for col in df.columns])))
    for demand in demands:
        df_demand = df[demand].T.reset_index()
        df_demand[df_demand["level_1"] == "mean"].plot(x="level_0", y=-1)
        plt.title(f"{metric} {demand}")
        plt.show()
```

```python
Toy_1000 = df["Toy_1000"].T.reset_index()
Toy_1000[Toy_1000["level_1"] == "mean"].plot(x="level_0", y=-1)
```
